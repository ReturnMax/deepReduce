```python
#csv_list保存所有文件的名字
```

```python
csv_list=[]
```

```python
for root,dirs,files in os.walk(wd):
    for file in files:
        if os.path.splitext(file)[1]==".csv":
            csv_list.append(file)
```

```python
data = []
```

```python
import time
for csv in csv_list:
    csv = csv_list[0]
    seqs=pd.read_csv(csv,header=None,encoding="utf8")
    seqs_list=seqs.values.flatten()

    gene_list="\t".join(seqs_list)

    data.append(gene_list)
```

```python
with open("/home/externalFolder/demo_final_item/all_data_1919/chr10.txt.","a+") as chr10_gene:
    for line in data:
        chr10_gene.write(line+"\n")
```
